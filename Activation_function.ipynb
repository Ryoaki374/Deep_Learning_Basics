```python
import numpy as np
import matplotlib.pyplot as plt

# 入力レンジの定義
x = np.linspace(-5, 5, 500)

# 各活性化関数を定義
sigmoid = 1 / (1 + np.exp(-x))
tanh_act = np.tanh(x)
relu = np.maximum(0, x)
leaky_relu = np.maximum(0.1 * x, x)
maxout = np.maximum(x, 0.5 * x + 1)  # Maxoutの例
elu = np.where(x >= 0, x, np.exp(x) - 1)  # ELU (α=1.0)

activations = [
    ("Sigmoid", sigmoid),
    ("tanh", tanh_act),
    ("ReLU", relu),
    ("Leaky ReLU", leaky_relu),
    ("Maxout", maxout),
    ("ELU", elu),
]

# 2行3列のサブプロットを作成
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.flatten()

for ax, (name, y) in zip(axes, activations):
    ax.plot(x, y)
    ax.axhline(0, color='gray', linewidth=0.8)
    ax.axvline(0, color='gray', linewidth=0.8)
    ax.set_title(name)
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.tight_layout()
plt.show()
```
